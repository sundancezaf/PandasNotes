```python
# Getting values from only selected columns
Cars2 = Cars[['name','mpg','am','cyl','wt','hp']]
Cars2.head()
```

```python
# Finding maximum value from a column
bestMPG = Cars2['mpg'].idxmax()
bestMPG
```

```python
# Using that maximum value and getting the row while 
# also showing just selected columns
# Cars2[bestMPG:bestMPG+1] -- you only want to show that one row so slice
# Then drop the columns you don't want
print("Most efficient car:")
Cars2[bestMPG:bestMPG+1].drop(columns=['hp','wt'])
```

```python
# Finding the mean of a column
theMean = Cars['wt'].mean()
theMean = round(theMean,2)
print(f"Average weight %.5s" % theMean)
```

```python
# How to filter 
# You want to show only the cars that weigh less than the average weight:
# Cars2[(Cars2['wt'] < theMean)]

# You don't want all the columns, drop those you don't want:
# .drop(columns=['hp','am','cyl','mpg'])

# Finally, sort the result:
# lightCars = lightCars.sort_values(by=['wt'])

lightCars = Cars2[(Cars2['wt'] < theMean)].drop(columns=['hp','am','cyl','mpg'])
lightCars = lightCars.sort_values(by=['wt'])
print("Lighter cars:")
lightCars
```

```python
# Show only the Mercedes cars
Mercedes =Cars2[Cars2['name'].str.contains('Merc', na=False)]
print("Mercedes cars horse power:")
Mercedes.drop(columns=['mpg','am','cyl','wt'])
```
